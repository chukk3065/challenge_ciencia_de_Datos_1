##Analisís de la tienda 1.


```python
vtas_tienda1 = tienda1['Precio'].sum()
vtas_tienda1
```

```python
envios_tienda1 = tienda1['Costo de envío'].sum()
envios_tienda1
```

```python
costo_vtas_total1 = vtas_tienda1 + envios_tienda1
print('El costos de ventas (Venta más envio) de la tienda 1 es de', costo_vtas_total1)
```

Adicional: agregar columna de conversión a dolares en precio y envio


```python
tienda1['Precio_dolares']  = tienda1['Precio'] / 3783.08
tienda1['Envio_dolares'] = tienda1['Costo de envío'] / 3783.08
```

```python
tienda1.head()
```

###Ingreso Total de la Tienda 1

```python
Total_ventas_Dlls = tienda1['Precio_dolares'].sum()
print('El total de ventas de la tienda 1 (en Dolares) es de: ', Total_ventas_Dlls.round(2))
Total_ventas_PC = tienda1['Precio'].sum()
print('El total de ventas de la tienda 1 (en Pesos Colombianos) es de: ', Total_ventas_PC.round(2))
```

###Costo promedio del envio en la Tienda 1

```python
Costo_DLLS =  tienda1['Envio_dolares'].sum() / len(tienda1['Envio_dolares'])
print('El costo promedio del envio (en Dolares) de la tienda 1 es de: ', Costo_DLLS.round(2))
```

```python
Costo_PC =  tienda1['Costo de envío'].sum() / len(tienda1['Costo de envío'])
print('El costo promedio del envio de la tienda 1 es de: ', Costo_PC.round(2))
```

###Ingreso VS costo de envio.


```python
monto_prom = Total_ventas_Dlls / len(tienda1['Precio_dolares'])
Rel = monto_prom/Costo_DLLS
print('La relacion entre el envio y el ingreso de cada venta es de aproximadamente: ', (Rel).round(2))
```

###Productos mas vendidos por categoria en la tienda 1

```python
Cat_prod = tienda1.groupby('Categoría del Producto')['Producto'].value_counts()
display(Cat_prod)
```

```python
Cat_prod_mas_vendidos = Cat_prod[Cat_prod > 55].sort_values(ascending=False)
display(Cat_prod_mas_vendidos)
```

###Calificación promedio de la tienda dada por clientes (satisfacción)

```python
C_prom = tienda1['Calificación'].mean()
print('La calificación promedio de la tienda 1 es de: ', C_prom)
```

nota: .mean() aparentemente es para sacar promedios

###Productos más y menos vendidos en la Tienda 1

```python
P_mas_vendidos = tienda1['Producto'].value_counts()
P_mas_vendidos = P_mas_vendidos[P_mas_vendidos > 55].sort_values(ascending=False)
P_menos_vendidos = tienda1['Producto'].value_counts()
P_menos_vendidos = P_menos_vendidos[P_menos_vendidos <= 35].sort_values(ascending=True)
display(P_menos_vendidos)
display(P_mas_vendidos)
```

```python
def obtener_top_productos(serie):
    max_ventas = serie.max()
    productos_top = serie[serie == max_ventas].index.tolist()  # lista con todos los empatados
    return productos_top, max_ventas

productos_top, ventas_top = obtener_top_productos(P_mas_vendidos)

print("Productos más vendidos:", productos_top)
print("Ventas:", ventas_top)

def obtener_menos_vendidos(serie):
    min_ventas = serie.min()
    productos_menos = serie[serie == min_ventas].index.tolist()  # lista con todos los empatados
    return productos_menos, min_ventas


productos_menos, ventas_menos = obtener_menos_vendidos(P_menos_vendidos)

print("Productos menos vendidos:", productos_menos)
print("Ventas:", ventas_menos)
```

###Reporte de ingresos por tienda

```python
import matplotlib.pyplot as plt

tiendas = Resultados['Tienda']
ingresos = Resultados['Ingreso total (dólares)']

# Colores diferentes por barra (puedes cambiar la paleta si gustas)
colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

plt.figure(figsize=(10, 6))
plt.barh(tiendas, ingresos, color=colores)

plt.title('Ingreso Total por Tienda (USD)')
plt.xlabel('Ingreso (USD)')
plt.ylabel('Tienda')
plt.grid(axis='x', linestyle='--', alpha=0.5)

# Mostrar valores exactos al final de cada barra
for i, valor in enumerate(ingresos):
    plt.text(valor + max(ingresos)*0.01, i, f"${valor:,.2f}")

plt.tight_layout()
plt.show()

```

###Productos mas y menos vendidos por cada tienda

```python
import matplotlib.pyplot as plt

tiendas = concentrado_ventas['Tienda']
ventas = concentrado_ventas['Ventas']
productos = concentrado_ventas['Producto(s) más vendido(s)']

# Colores distintos para cada barra
colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

plt.figure(figsize=(10, 5))
plt.bar(tiendas, ventas, color=colores)

plt.title('Ventas del Producto Más Vendido por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Cantidad de Ventas')
plt.grid(axis='y', linestyle='--', alpha=0.4)

# Ajustar escala del eje y a 0-100
plt.ylim(0, 100)

# Mostrar producto y valor encima de cada barra
for i, (venta, producto) in enumerate(zip(ventas, productos)):
    plt.text(i, venta + 1,  # un poco por encima de la barra
             f"{producto}\n({venta})",
             ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Función para obtener los menos vendidos
def obtener_menos_vendidos(serie):
    min_ventas = serie.min()
    productos_menos = serie[serie == min_ventas].index.tolist()  # lista con todos los empatados
    return productos_menos, min_ventas

# Obtener los menos vendidos de cada tienda
productos_menos_vendidos, ventas_menos_vendidos = [], []

for P_menos in [P_menos_vendidos, P_menos_vendidos2, P_menos_vendidos3, P_menos_vendidos4]:
    productos, ventas = obtener_menos_vendidos(P_menos)
    # Concatenar los nombres si hay empate
    productos_menos_vendidos.append(', '.join(productos))
    ventas_menos_vendidos.append(ventas)

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Colores para cada barra
colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

plt.figure(figsize=(12, 5))
plt.bar(tiendas, ventas_menos_vendidos, color=colores)

plt.title('Ventas del Producto Menos Vendido por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Cantidad de Ventas')
plt.grid(axis='y', linestyle='--', alpha=0.4)

# Ajustar escala del eje y
plt.ylim(0, max(ventas_menos_vendidos)*1.2)

# Mostrar producto(s) y valor encima de cada barra
for i, (venta, productos) in enumerate(zip(ventas_menos_vendidos, productos_menos_vendidos)):
    plt.text(i, venta + 0.5, f"{productos}\n({venta})", ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
